# Root mean square deviation

It is often useful to quantify the geometrical similarity between two structures. For instance, when simulating a solid at finite temperature, you may want to give a number to the difference between the sample at equilibrium and the sample undergoing vibration.

A common measure of geometrical similarity is the **Root-Mean-Square Deviation (RMSD)**. The formula for the RMSD between two structures $A$ and $B$ of $N$ atoms, each is:
$$
RMSD = \sqrt{\frac{\sum_i^N (x_{A,i} - x_{B,i})^2 + (y_{A,i} - y_{B,i})^2 + (z_{A,i} - z_{B,i})^2}{N}}
$$
Where $x_{A,i}$ is the $x$ coordinate of atom $i$ in molecule $A$, and so on.

# File format

In Computational Chemistry, structures are often stored in so-called `.xyz` file formats. A `.xyz` file for water looks like this:
```
3
water
O    0.000  0.000  0.000
H    0.000  0.762 -0.599
H    0.000 -0.762 -0.599
```

- Line 1 states how many atoms are in the file.
- Line 2 is a free line where the user can write whatever they want.
- Lines 3+ state the elements and Cartesian coordinates of each atom, separated by white spaces.

You can visualise the atoms inside an `.xyz` file by using programs such as Avogadro or VESTA, or by using an [online viewer](http://calistry.org/calculate/xyzviewer).

To write an `.xyz` file, simply write the contents in a text editor like Spyder, VSCode, or notepad, and save the file with the name ending in `.xyz`.
