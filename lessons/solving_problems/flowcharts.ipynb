# Flowcharts

A lesson on how to use flowcharts to visualise and represent your problem.

If you are stuck on understanding how to iterate through a loop or use an if statement, you can use a flowchart to help you visualise the problem. They are also good to help others understand what your program is achieving.

## Guide to flowcharts and their symbols

All programming flowcharts use a similar system of symbols to make them easier to comprehend.
- Oval boxes indicate the start or end.
- Rhombus boxes represent an input or output.
- Square boxes show a process, e.g. a calculation. 
- Diamond indicates a decision.

All follow the same rules:

- All must have a start and end point.
- The arrows must be single-headed and show progression. The arrows must start at the start point and end at an end-point.
