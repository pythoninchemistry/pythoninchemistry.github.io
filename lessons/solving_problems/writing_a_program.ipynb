## Writing a program to calculate the mean and standard deviation of a set of numbers

### Establishing basics

Now we can pull all we have learnt and write our own program.

We will create a program to calculate the mean and standard deviation of a set of numbers. We will therefore need the following equations:

For the mean: $$<x> = \frac{1}{N}(x_1 + x_2 + x_3 + ...+x_n)= \frac{1}{N}\sum(x_i) $$

So the mean is the sum of N numbers from $x_1$ to $x_n$

The Population has variance: $$ v(x,x)=<(x-\bar{x})^2> = \frac{1}{N}\sum((x_i -\bar{x})^2) $$

The variance is the sum of the <b>squared </b> differences of N numbers from $x_1$ to $x_n$

For the standard deviation: $$\sigma(\bar{x}) = \sqrt{v(x,x)}$$

The standard deviation is the square root of the variance.

We will take this equations and turn them into a working programme
